## Сельское хозяйство

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Показатели, характеризующие импортозамещение в России

#### Производство основных продуктов растениеводства

В расчёты взяты хозяйства всех категорий

```python
main_crop = pd.read_excel("C:/Users/larag/Downloads/ПРОИЗВОДСТВО ОСНОВНЫХ ПРОДУКТОВ РАСТЕНИЕВОДСТВА.xls", sheet_name='1')
main_crop.index = ['Kартофель', 'Овощи', 'Плоды и ягоды']
main_crop = main_crop.drop(columns=main_crop.columns[0])
main_crop = main_crop.T.round(1)
main_crop['Все категории'] = main_crop['Kартофель'] + main_crop['Овощи'] + main_crop['Плоды и ягоды']
main_crop
```

```python
potato_line = main_crop['Kартофель'].median()
vegetables_line = main_crop['Овощи'].median()
berries_line = main_crop['Плоды и ягоды'].median()
total_line = main_crop['Все категории'].median()
print(potato_line, vegetables_line, berries_line, total_line)
```

```python
main_crop.plot(grid=True)
plt.title('Графики производства основных продуктов растениеводства')
plt.ylabel('Урожай, тыс. тонн')
plt.xlabel('Годы')
plt.axhline(y=potato_line, color='blue', linestyle='--')
plt.axhline(y=vegetables_line, color='orange', linestyle='--')
plt.axhline(y=berries_line, color='green', linestyle='--')
plt.axhline(y=total_line, color='red', linestyle='--')
None
```

*Промежуточный вывод*

Нами рассматривалось производство основных продуктов растениеводства (картофель, овощи, ягоды и плоды) за 2010 - 2021 года. В 2011 году выросла урожайность всех рассматриваемых категорий. 

Урожайность картофеля показывает отрицательную динамику с 2011 года, к 2021 году отставая от показателей 2011 года (наилучший показатель) на 36% и в 2016 году опустилась ниже уровня медианы. 

Урожайность овощей возрастала с 2013 года, к 2016 году преодолев порог медианы, с 2019 года наблюдается незначительный спад урожайности.

Урожайность ягод показывает положительную динамику. К 2021 году показатели улучшились вдвое по отношению к 2010 году

Итог:

Наибольшая по урожайности (медиана 22429 тыс. тонн) рассматриваемая категория - картофель, демонстрирует продолжительный спад, следующая по урожайности (медиана 13182 тыс. тонн) категория - овощи, показывая отрицательную динамику с 2019 года, оставаясь выше уровня медианы, наименьшая по урожайности (медиана 2759 тыс. тонн) категория плодов и ягод на всём рассматриваемом промежутке демонстрирует положительную динамику. Общие показатели отрасли снижаются и находятся нижу уровня медианы с 2015 года.

#### Производство основных продуктов животноводства

В рассчёты взяты хозяйства всех категорий

```python
animals_stat = pd.read_excel("C:/Users/larag/Downloads/животноводство.xls", sheet_name='животноводство')
animals_stat.index=['Скот и птица на убой (в живом весе)', 'Крупный рогатый скот', 'Свиньи', 'Птица', 'Молоко']
animals_stat = animals_stat.T.round(1)
animals_stat
```

```python
cattle_line = animals_stat['Крупный рогатый скот'].median()
pigs_line = animals_stat['Свиньи'].median()
poultry_line = animals_stat['Птица'].median()
milk_line = animals_stat['Молоко'].median()
print(cattle_line, pigs_line, poultry_line, milk_line)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

plot1 = animals_stat[['Крупный рогатый скот', 'Свиньи', 'Птица']].plot(grid=True, ax=axes[0])
axes[0].set_title('Графики производства основных продуктов животноводства')
axes[0].set_ylabel('Общий вес, тыс. тонн')
axes[0].set_xlabel('Годы')
axes[0].axhline(y=cattle_line, color='brown', linestyle='--')
axes[0].axhline(y=pigs_line, color='orange', linestyle='--')
axes[0].axhline(y=poultry_line, color='green', linestyle='--')

plot1.lines[0].set_color('brown')

milk_table = animals_stat['Молоко'].plot(grid=True, ax=axes[1])
axes[1].set_title('График производства основных продуктов животноводства (молоко)')
axes[1].set_ylabel('Общий вес, тыс. тонн')
axes[1].set_xlabel('Годы')
axes[1].axhline(y=milk_line, color='blue', linestyle='--')

milk_table.legend(['Молоко', 'Медиана'])

plt.show()
```

*Промежуточный вывод*

Нами рассматривалось производство основных продуктов животноводства: скот и птица на убой (в живом весе), крупный рогатый скот, свиньи, птица, молоко. 

Динамика производства птицы показывает уверенный рост (на 81% с 2010 года), в 2016 году преодолевая порог медианы. С 2017 года темпы роста замедлились (к 2022 году рост на 6% по сравнению с 2017 годом)

Динамика производства свинины так же демонстрирует уверенный рост (на 86% по сравнению с 2010 годом), в 2016 году преодолевая порог медианы и сохряняя темпы роста до 2022 года

В 2011 году производство крупного рогатого скота сниажется 27% по сравнению с 2010 годом и колеблется относительно медианы в течение всего периода наблюдений

В 2010 - 2016 годах наблюдался спад производства молока, с 2016 года ситуация улучшалась, к 2019 преодолев порог медианы. К 2022 году показатели демонстрируют рост на 4% относительно показателей 2010 года и на 10% относительно показателей 2016 года

Итог:

Наибольшие по производству отрасль животноводства - производство молока - (медиана 31196 тыс. тонн) восстановилась после спада 2010-2016 годов, следующие по объёмам производства отрасли - производство птицы и свиньи (медианы 6390 и 4329 тыс. тонн соответственно) - показывают уверенный рост, производство крупного рогатого скота остаётся стабильным на протяжении всего периода наблюдений

### Себестоимость производства и реализации основных видов сельскохозяйственной продукции в сельскохозяйственных организациях

#### Уровень инфляции

```python
inflation = pd.read_excel("C:/Users/larag/OneDrive/Документы/Инфляция.xlsx", index_col=0)
inflation = inflation[::-1]
inflation
```

#### Производственная себестоимость продукции растениеводства

```python
cost_crop = pd.read_excel("C:/Users/larag/Downloads/cost_crop.xls", sheet_name='Данные')
cost_crop = cost_crop.T
cost_crop.index.name = 'Год'
cost_crop.columns = ['Стоимость']
cost_crop
```

```python
concatenated_df = pd.concat([cost_crop, inflation], axis=1)
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title('Производственная себестоимость продукции растениеводства')
axes[0].set_ylabel('Себестоимость, тыс. руб.')
concatenated_df['Стоимость'].plot(grid=True, ax=plt.subplot(1, 2, 1))
concatenated_df['Инфляция'].plot(grid=True, ax=plt.subplot(1, 2, 2))
axes[1].set_title('Инфляция')
axes[1].set_ylabel('Инфляция, %')
None
```

*Промежуточный вывод*

Производственная себестоимость продукции растениеводства к 2022 году возросла на 336%, когда уровень инфляции увеличился на 111% по состоянию на соответствующий промежуток времени. Также снижение уровня инфляции не вызывало существенного замедления роста себестоимости продукции

#### Производственная себестоимость продукции животноводства

```python
cost_animals = pd.read_excel("C:/Users/larag/Downloads/cost_animals.xls", sheet_name='Данные')
cost_animals = cost_animals.T
cost_animals.index.name = 'Год'
cost_animals.columns = ['Стоимость']
cost_animals
```

```python
concatenated_df = pd.concat([cost_animals, inflation], axis=1)
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].set_title('Производственная себестоимость продукции животноводства')
axes[0].set_ylabel('Себестоимость, тыс. руб.')
concatenated_df['Стоимость'].plot(grid=True, ax=plt.subplot(1, 2, 1))
concatenated_df['Инфляция'].plot(grid=True, ax=plt.subplot(1, 2, 2))
axes[1].set_title('Инфляция')
axes[1].set_ylabel('Инфляция, %')
None
```

*Промежуточный вывод*
